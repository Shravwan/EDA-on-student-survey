```python
import pandas as pd

# Load the CSV file
file_path = '/content/SASTRA HOSTEL SURVEY (Responses) - Form Responses 1.csv'
data = pd.read_csv(file_path)
```

```python
# Display column names and the first few rows
print(data.columns)
print(data.head())
```

```python
# Columns to drop (make sure the column names match exactly)
columns_to_drop = [
    'Timestamp', 'Name:-', 'Place of Origin:', "Father's Occupation:", "Mother's Occupation:",
    'Annual Salary of Family', 'Degree Programme:-', 'Year of Study:-', 'Branch:-', 'Hostel Name & Room Type: '
]

# Drop the columns
cleaned_data = data.drop(columns=columns_to_drop)
```

```python
cleaned_data.head()
```

```python
# Drop rows with more than 50% missing data
cleaned_data = cleaned_data.dropna(thresh=int(cleaned_data.shape[1] * 0.5))
```

```python
cleaned_data.head()
```

```python
# Save the cleaned data to a new CSV file
cleaned_data.to_csv('cleaned_survey_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/Parents_ResponseV2.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)


# Remove the last column
df = df.iloc[:, :-1]

# Rename the columns
df.columns = [
    'Hostel_Rent_Comparison_Private',
    'Comparable_Hostel_Accommodation',
    'Hostel_Rent_Comparison_Universities',
    'Hostel_Security_Comparison',
    'Room_Comfort_Comparison',
    'Food_Quality_Comparison',
    'Food_Variety_Comparison',
    'Room_Cleanliness_Comparison',
    'Toilet_Cleanliness_Comparison',
    'Hot_Water_Geyser_Comparison',
    'Cooking_Utilities_Comparison',
    'Support_Facilities_Comparison',
    'Entertainment_Sports_Comparison',
    'WiFi_Power_Comparison',
    'Room_Infra_Comparison',
    'Hostel_Rules_Comparison',
    'Hostel_Rent_Comparison',
    'Mess_Fee_Comparison',
    'Water_Availability_Comparison',
    'Room_Mate_Comparison',
    'Support_Staff_Comparison'
]

# List of columns to normalize
columns_to_normalize2 = [
     'Comparable_Hostel_Accommodation' # Replace with your actual column names
]

for col in columns_to_normalize2:
    df[col] = df[col].str.strip().str.title()

    # If you want to replace specific variations, you can do so here
    df[col] = df[col].replace({
        'Yes, comparable':'Yes, Comparable',
        'Yes,Comparable':'Yes, Comparable',
        'No, outside is better': 'No, Outside Is Better'
        # Add more mappings as needed
    })


columns_to_normalize = [
    'Hostel_Security_Comparison',
    'Room_Comfort_Comparison',
    'Food_Quality_Comparison',
    'Food_Variety_Comparison',
    'Room_Cleanliness_Comparison',
    'Toilet_Cleanliness_Comparison',
    'Hot_Water_Geyser_Comparison',
    'Cooking_Utilities_Comparison',
    'Support_Facilities_Comparison',
    'Entertainment_Sports_Comparison',
    'WiFi_Power_Comparison',
    'Room_Infra_Comparison',
    'Hostel_Rules_Comparison',
    'Hostel_Rent_Comparison',
    'Mess_Fee_Comparison',
    'Water_Availability_Comparison',
    'Room_Mate_Comparison',
    'Support_Staff_Comparison'  # Replace with your actual column names
]

for col in columns_to_normalize:
    df[col] = df[col].str.strip().str.title()

    # If you want to replace specific variations, you can do so here
    df[col] = df[col].replace({
        'better': 'Better',
        'bettter':'Better',
        'Bettter':'Better',
        'Better':'Better',
        'Bettet':'Better',
        'same': 'Same',
        'inferior': 'Inferior',
        'low':'Inferior',
        'Low':'Inferior'
        # Add more mappings as needed
    })

# Basic EDA

# 1. Check for missing values
missing_values = df.isnull().sum()

# 2. Summary statistics
summary_stats = df.describe()

# 3. Distribution of categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.show()
#3.1 Pie chart for the categorical variables
for col in categorical_columns:
    plt.figure(figsize=(8, 8))
    df[col].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')  # Hides the y-label
    plt.show()
# 4. Correlation matrix (if numeric data is present)
if not df.select_dtypes(include=['number']).empty:
    correlation_matrix = df.corr()
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()


# Output the results
print("Missing Values:\n", missing_values)
print("\nSummary Statistics:\n", summary_stats)
```
